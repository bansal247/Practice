```python
import win32api

drives = win32api.GetLogicalDriveStrings()
drives = drives.split('\000')[:-1]
print(drives)
```

```python
file_path_list = []
def scan_path(path):
    try:
        os.chdir(path)
        files = os.listdir()
    except Exception as e:
        return
    for file in files:
        if '.' in file:
            file_name = file.split('.')[0]
            ext_name = file.split('.')[1]
            if file_name.lower().find(f_name.lower()) > -1 and ext_name == f_exten:
                file_path_list.append(path+file)
        else:
            scan_path(path+'{}\\'.format(file))
```

```python
for i in drives:
    scan_path(i)
```
